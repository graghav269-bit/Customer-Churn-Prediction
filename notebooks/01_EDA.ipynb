Dataset Overview

```python
import pandas as pd

df = pd.read_csv("../data/customer_churn.csv")

df.head()
```

Missing Values

```python
df.isnull().sum()
```

Churn Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(
    data=df,
    x="Churn"
)

plt.show()
```

Contract Analysis

```python
sns.countplot(
    data=df,
    x="Contract",
    hue="Churn"
)

plt.xticks(rotation=20)
plt.show()
```

Monthly Charges Analysis

```python
sns.boxplot(
    data=df,
    x="Churn",
    y="MonthlyCharges"
)

plt.show()
```

Tenure Analysis

```python
sns.boxplot(
    data=df,
    x="Churn",
    y="tenure"
)

plt.show()
```

Correlation Matrix

```python
numeric_df = df.select_dtypes(
    include=["int64", "float64"]
)

sns.heatmap(
    numeric_df.corr(),
    annot=True
)

plt.show()
```
